```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('./Dataset/fods_1.csv')
print(len(df))
```

```python
def normalise(X):
    mean = np.mean(X,axis=0)
    sd = np.std(X,axis=0)
    X = (X-mean)/sd
    return X
    
```

```python
df.head()
```

```python
df = df.sample(frac=1)
```

```python
length = (df.shape[0] * 0.8)
length = int(length)
X_train = normalise(np.array(df.iloc[0:length,0:2]))
Y_train = normalise(np.array(df.iloc[0:length,[2]]))
# X_test = normalise(np.array(df.iloc[length:df.shape[0],0:2]))
# Y_test = normalise(np.array(df.iloc[length:df.shape[0],[2]]))
X_train = X_train.T
Y_train = Y_train.T
```

```python
def grad_desc(X,Y,lr=0.01,epoch=1000):
    J = []
    w1 = []
    w2 = []
    w = np.random.randn(1,2)
    for i in range (epoch):
        Yhat = np.dot(w,X)
        loss = np.average((Yhat - Y) ** 2) / 2
        J.append(loss)
        w1.append(w[0][0])
        w2.append(w[0][1])
        X1 = X[:,0]
        X2 = X[:,1]
        grad1 = np.average( np.multiply(Yhat-Y,X[0]) )
        grad2 = np.average(np.multiply(Yhat-Y,X[1]))
        w[0][0] = w[0][0] - lr*grad1
        w[0][1] = w[0][1] - lr*grad2
    return J,w1,w2
```

```python
J_1,w11,w12 = grad_desc(X_train,Y_train)
```

```python
plt.plot(J_1)
```

```python
J1 = np.array(J_1)
J1[999] # OPtimal J value
```

```python
def cost_func(w1,w2):
    m = X_train.shape[1]
    ans  = 0
    for i in range(m):
        temp = (w1*X_train[0][i] + w2*X_train[1][i] - Y_train[0][i]) ** 2
        ans = ans + temp
    ans = ans / (2*m)
    return ans
```

```python
w11 = np.array(w11).reshape(len(w11),1)
w12 = np.array(w12).reshape(len(w12),1)
w12[999]
```

```python
x = np.arange(-4,4,0.005)
y = np.arange(-4,4,0.005)

X, Y = np.meshgrid(x, y)
Z = cost_func(X, Y)
```
