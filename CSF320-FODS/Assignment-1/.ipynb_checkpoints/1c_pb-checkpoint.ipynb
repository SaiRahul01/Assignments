```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./Dataset/fods_1.csv')
print(len(df))
```

```python
def normalise(X):
    mean = np.mean(X,axis=0)
    sd = np.std(X,axis=0)
    X = (X-mean)/sd
    return X
    
```

```python
df.head()
```

```python
df = df.sample(frac=1)
```

```python
length = (df.shape[0])
length = int(length)
X_train = normalise(np.array(df.iloc[0:length,0:2]))
Y_train = normalise(np.array(df.iloc[0:length,[2]]))
X_test = X_train #normalise(np.array(df.iloc[length:df.shape[0],0:2]))
Y_test = Y_train #normalise(np.array(df.iloc[length:df.shape[0],[2]]))
X_train = X_train.T
Y_train = Y_train.T
X_test=X_test.T
Y_test=Y_test.T
```

```python
def grad_desc(X,Y,lr=0.01,epoch=1000):
    J = []
    w1 = []
    w2 = []
    w = np.zeros((1,2))
    for i in range (epoch):
        Yhat = np.dot(w,X)
        loss = np.average((Yhat - Y) ** 2) / 2
        J.append(loss)
        w1.append(w[0][0])
        w2.append(w[0][1])
        grad1 = np.average( np.multiply(Yhat-Y,X[0]))
        grad2 = np.average(np.multiply(Yhat-Y,X[1]))
        w[0][0] = w[0][0] - lr*grad1
        w[0][1] = w[0][1] - lr*grad2
    return J,w1,w2
```

```python
J_1,w11,w12 = grad_desc(X_train,Y_train)
```

```python
def cost_func(w1,w2):
    m = X_train.shape[1]
    ans_lst=[]
    for j in range(w1.shape[0]):
        temp=np.average((w1[j]*X_train[0,:]+w2[j]*X_train[1,:]-Y_train[0,:])**2)/2
        ans_lst.append(temp)
    return np.array(ans_lst).reshape(len(ans_lst),1)
```

```python
w11 = np.array(w11).reshape(len(w11),1)
w12 = np.array(w12).reshape(len(w12),1)
w12[999]
```

```python
x = np.arange(-2,2,0.005)
y = np.arange(-2,2,0.005)
X, Y = np.meshgrid(x, y)
print(X.shape)
print(Y.shape)

Z = cost_func(X.flatten(), Y.flatten()).reshape(X.shape[0],X.shape[1])
#X.flatten().shape
Y_train.shape
```

```python
def in_range(eeta,q,delta=0.005):
    points1 = []
    points2 = []
    for i in range(len(w11)):
        if abs(w11[len(w11)-i-1])**q+abs(w12[len(w11)-i-1])**q <= eeta+delta and abs(w11[len(w11)-i-1])**q+abs(w12[len(w11)-i-1])**q >= eeta-delta:
            points1.append(w11[len(w11)-i-1])
            points2.append(w12[len(w11)-i-1])
    return points1,points2
```

```python
plt.figure(figsize=(6,6))
plt.contour(X, Y, Z, colors='black',levels=100)
plt.title("Regularized Cost Contour(q=0.5)")
plt.xlabel("W1")
plt.ylabel("W2")
l1 = np.arange(-2,2,0.05)
l2 = (1.4 - (abs(l1) ** 0.5))**2
plt.plot(l1,l2,'g')
l1 = np.arange(-2,2,0.05)
l2 = -1 * ((1.4 - (abs(l1) ** 0.5))**2)
plt.plot(l1,l2,'g')
temp1=np.arange(-2,w11[999],0.05)
temp2=np.arange(-2,w12[999],0.05)
plt.plot(temp1,np.ones(len(temp1))*w12[999],"-b",linestyle="dashed")
plt.plot(np.ones(len(temp2))*w11[999],temp2,"-r",linestyle="dashed")
plt.plot(w11,w12,linewidth=2)
plt.scatter([0],[1.957],color='#FF0000',marker='X')
plt.legend(['Constraint','Constraint','W2 Unregularized','W1 Unregularized','tranisition of weight','Regularized weight'],loc='lower left')
```
